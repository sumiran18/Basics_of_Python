# **The Control Statement**

```python
number=23
guess=int(input("Enter the number"))
# input(), by default takes the input in string
if guess==number:
  #if block starts here
  print("You guessed the number right")
  #if block ends here
elif guess<number:
  #elif block start here
  print("You guessed a number lower than the original number")
else:
  print("You guessed a number higher than the original number")
```

```python
number=23

running=True
#True and False are called the Boolean Types , equivalent to 0 and 1
while running:
  guess=int(input("Enter a number"))
  if guess==number:
    print("You guessed the number right")
    running=False
  elif guess<number:
    print("You guessed a number lower than the original number")
  else:
    print("You guessed a number higher than the original number")
else:
  print("The loop is over")
# you can have an else clause for while loop
```

# **Tuples**

```python
#use of parenthesis is optional
#explicit is always better than implicit
zoo=('python','elephant','lion')
#tuples are immutable , it cannot be altered
print("The Number of animals in the zoo is:",len(zoo))
new_zoo=('monkey','penguin',zoo)
#use of parenthesis is optional
print('\nThe Number of cages in the new zoo is:', len(new_zoo))
print('\nAll the animals in the new zoo are:',new_zoo)
print('\nAnimals bought from the old zoo are:', new_zoo[2])
#new_zoo[2]--> indexing operator
print('\nLast animal bought from old zoo is:', new_zoo[2][2])
#tuple within a tuple does not lose its identity
print('\nThe number of animals in the new zoo is:', len(new_zoo)- 1 + len(new_zoo[2]))
```

# **Dictionary**

```python
ab={
    'Sumiran':'sumirangrover.1823@gmail.com',
    'Seema': 'seemagrover2803@gmail.com',
    'Rishi': 'rishigrover22@gmail.com',
    'Dhairya': 'dhairyagrover31@gmail.com',
    'Spammer': 'spammer@hotmail.com'
}
print("Sumiran's address is:", ab['Sumiran'])

#deleting a key value pair
del ab['Spammer']
print('\nThere are {} contacts in the address book:'.format(len(ab)))
for name,address in ab.items():
#items() is a method of dictionary , which returns a tuple
  print('\nThe contact of {} is {}'.format(name,address))
#adding a key value pair
ab['Sumiran']='sgrover_be@thapar.edu'
if 'Sumiran'in ab:
#we can check if a key-value pair exists using the "in" operator
  print("\nSumiran's address is:",ab['Sumiran'])
```

# **Sequences**

```python
shoplist=['Apple', 'Mango','Grapes', 'Watermelon']
name="Sumiran"

#indexing or subscription operation
print('\nItem[0] is:', shoplist[0])
print('\nItem[1] is:', shoplist[1])
print('\nItem[2] is:', shoplist[2])
print('\nItem[-1] is:', shoplist[-1])
print('\nItem[-2] is:', shoplist[-2])
print('\nCharater 0 is:',name[0])

#negative index--> postion is calculated from the end of the list
#shoplist[-1] --> refers to the last item in the list
```

```python
#slicing operation in a string
print('\nCharacter 1 to 3 is:',name[1:3])
print('\nCharacter 2 to end is:',name[2:])
print('\nCharacter 1 to -1 is:',name[1:-1])
print('\nCharacter from start to end is:',name[:])
#first number--> where the slice starts
#second number--> where the slice will stop at
#if the first number is not specified, it will start from the beginning of sequence
#if the secon number is not specfied, it will stop at the end of the seqeunce
#name[1:3]--> returns the slice of sequence starting at postion 1, including 2 and excluding 3
#start postion is included
#end position is excluded
#you can also specify a third argument i.e. the slice count
#by default, the slice count is 1
```

```python
#slicing operation in a list
print('\nItem 1 to 3 is:',shoplist[1:3])
print('\nItem 2 to end is:',shoplist[2:])
print('\nItem from start to end is:',shoplist[:])
print('\nItem 1 to -1 is:',shoplist[1:-1])
```
